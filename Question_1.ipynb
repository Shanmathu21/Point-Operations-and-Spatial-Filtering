## 1. Intensity Transformation

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load grayscale image
Emma = cv.imread('a1images/emma.jpg', cv.IMREAD_GRAYSCALE)
assert Emma is not None

# Define transformation function
# Maps input 0–50 → output 0–50
t1 = np.linspace(0, 50, num=51).astype('uint8')
# Maps input 51–150 → output 100–255
t2 = np.linspace(100, 255, num=100).astype('uint8')
# Maps input 151–255 → output 150–255
t3 = np.linspace(150, 255, num=105).astype('uint8')

# Concatenate all segments to create the transformation array
t = np.concatenate((t1, t2, t3), axis=0).astype('uint8')
print(t.shape)

# Plot the array
plt.figure(figsize=(5, 5))
plt.plot(np.arange(256), t, color='blue', linewidth=2)
plt.title("Transformation Curve")
plt.xlabel("Input Intensity")
plt.ylabel("Output Intensity")
plt.xlim(0, 255)
plt.ylim(0, 255)
plt.grid(True)
plt.show()
```

```python
# Apply transformation 
g = t[Emma] # Map each pixel

# Display the image
fig, ax = plt.subplots(1, 2, figsize=(8, 6))
ax[0].imshow(Emma, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].imshow(g, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Transformed')
ax[1].axis('off')

plt.tight_layout()
plt.show()
```

# Approach


The given transformation curve was applied to the grayscale version of the image, boosting mid-gray intensities while keeping shadows and highlights mostly unchanged. The mapping was implemented pixel-wise according to the curve, and both the transformation plot and processed image were produced for verification.

# Interpretation

This operation brightens and enhances contrast in mid-tone regions, making details more noticeable, while leaving dark and bright areas largely unaffected. The abrupt changes in the curve create sharper transitions that can enhance edges but may also introduce visible intensity jumps in smooth gradients.
